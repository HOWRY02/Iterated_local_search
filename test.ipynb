```python
def find_inventory_levels(customers, forecasted_quantities):
    inventory_levels = np.ones_like(forecasted_quantities)
    for i, forecasted_quantity in forecasted_quantities:
        current_value = customers[i].init_quantity
        for j, value in forecasted_quantity:
            inventory_levels[i][j] = current_value - (value/customers[i].capacity)*100
            current_value = inventory_levels[i][j]

    return inventory_levels
```

```python
abc = [[[1, 2], []], [[4, 5, 6], [4, 5, 6]]]

def create_zero_list(data):
  """Creates a list with the same shape as data filled with zeros."""
  zero_list = []
  for sublist1 in data:
    inner_list = []
    for sublist2 in sublist1:
      inner_list.append([0] * len(sublist2))  # Create a list with zeros for each sub-sublist
    zero_list.append(inner_list)
  return zero_list

zero_abc = create_zero_list(abc)
print(zero_abc)
```
